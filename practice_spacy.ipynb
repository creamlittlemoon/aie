```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

# POS tagging and NER
POS: part-of-speech

```python
text = "Apple Inc. reported a 10% increase in its quarterly revenue, while Microsoft Corp. announced a major acquisition of LinkedIn for $26.2 billion."
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

# extract important parts

```python
text = "Apple Inc. reported a 10% increase in its quarterly revenue, while Microsoft Corp. announced a major acquisition of LinkedIn for $26.2 billion."
doc = nlp(text)
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}, Root: {ent.root.head.text}")
keywords = ["revenue", "acquisition"]
for word in keywords:
    print(f"Words similar to '{word}':")
    for token in nlp.vocab:
        if token.has_vector and token.is_lower and token.is_alpha:
            similarity = nlp(word).similarity(token)
            if similarity > 0.7:  # Adjust threshold as needed
                print(f"  {token.text}: {similarity:.4f}")
```

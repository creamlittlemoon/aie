```python
from typing import List, Tuple

def word_count_order(word_list: List[str]) -> List[Tuple[str, int]]:
    count = {}
    order = []

    for w in word_list:
        if w not in count:
            order.append(w)
        count[w] = count.setdefault(w, 0) + 1

    return sorted(
        [(w, count[w]) for w in order],
        key=lambda x: -x[1]
    )
```

```python
# self 20260127 ‚ùå

def longest(s_string:str)->int:
    # sliding window
    left = 0
    last = {}   #record the word value and position we've seen, so it's this pair -> word:position
    best = 0    #store the best/longest substring, which will be updated on condition
    
    # current window = [left, right], inside is the characters
    for right, ch in enumerate(s_string):
        # determine whether ch is in current window
        if ch in last and last[ch] >= left:
            left = last[ch] + 1 #‚ö†Ô∏è update left only when we don't need the old left boundary any moreÔºå and we just simply set the left boundary to the next word of the old word
        last[ch] = right    # ‚ö†Ô∏è everytime we should update last[ch]!!
            ## WRONG below...
            # if in current window, we should update our left, and the last seen dict
            # left = right + 1
            # last[ch] = right 
        best = max(best, right - left + 1)

    return best

s = "abcabcbb"
longest(s)
```

```python
"""  
EXPLANATION
1.  last[ch] >= left
        for example, "abba".
        when we process the second b, we already tossed the first a (index=0), and made left=last[ch]+1=1+1=2
        which means, when we process the second a(index=2), the old a(index=0) doesn't matter any more because the start of the window, which is the "left", is already 2!
"""

def longest_unique_substring_len(s: str) -> int:
    """ Keep a window that ALWAYS contains UNIQUE characters, expand it to the right, 
    and shrink it from the left when a duplicate appears. """
    last = {}   # a dict to keep: character ‚Üí last index where it appeared
    left = 0    # The start index of the current valid window / the earliest position of the valid substring
    best = 0    # stores the max length we've found so far
    for right, ch in enumerate(s):
        if ch in last and last[ch] >= left:     #MOST important line! If the previous ch is outside the window, it does NOT cause a conflict
            left = last[ch] + 1     # use 'abc a' as an example, here >= left means: The earliest valid substring must start after the old 'a'
        last[ch] = right    # update where we last saw this character
        best = max(best, right - left + 1)  # right-left+1: current window length
    return best

s = "abcabcbb"
print(longest_unique_substring_len(s))
```

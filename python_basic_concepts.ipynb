###  ‚ö†Ô∏èedge case: always use DEFAULT! 

```python
# check where the exact word is in

from collections import defaultdict

docs = {
  1: "refund policy within 7 days",
  2: "shipping takes 3-5 business days",
  3: "refund requires receipt"
}

index = defaultdict(set)    # creates a dict where missing keys start as empty sets.
for doc_id, text in docs.items():
    for w in text.split():  #split the text into words
        index[w.lower()].add(doc_id)    #For each word, lowercase it and add the doc_id to index[word]

def retrieve(q: str):
    return sorted(index.get(q.lower(), []))     #it no appearance then return []. ‚ö†Ô∏èif no DEFAULT=[] then error

print(retrieve("Love"))

"""  
This pattern appears in:
    Keyword search
    Hybrid RAG (BM25 + embeddings)
    Candidate recall stages

"""
```

## setdefault(): grouping (bucket)
(strong interview signal)
- mental model: grouping = create BUCKETS, then fill them
- RL intuition

```python
logs = [
    {"user": "A", "reward": 1},
    {"user": "A", "reward": 0},
    {"user": "B", "reward": 1}
]

group = {}
for log in logs:
    u = log["user"]
    group.setdefault(u,[]).append(log["reward"])
    """ setdefault()
    If u is already a key, return group[u].
    If u is missing, create group[u] = [] and return that list.
    """
group
```

## defaultdict
from collections import defaultdict
- it's used when you want a dictionary that automatically creates default values for missing keys, instead of raising a KeyError

```python
# without defaultdict, we have to check every kv
# for example

# we want to do this, but dict d doesn't have key a, so it will be KeyError
d = {}
d['a'].append(1)   # KeyError

# so we have to guard every access like this:
if 'a' not in d:
    d['a'] = []
d['a'].append(1)
```

```python
# example: grouping values by key

from collections import defaultdict

logs = [
    ("user1", "login"),
    ("user2", "pay"),
    ("user1", "logout")
]

events = defaultdict(list)

for user, action in logs:
    events[user].append(action)

print(events)
```

## yield
- return: ends the function/loop immediately
- yield: when we want usable pieces, use yield. ü§©
- print: only print TEXTs ‚ö†Ô∏è

```python
def simple_generator(n):
    i = 0
    while i < n:
        yield i
        i += 1

# Using the generator function
gen = simple_generator(5)
for number in gen:
    print(number)
```

# sliding window
easy to medium

```python
# Problem: Given daily delivery times, find the minimum average time over any consecutive window of size k.

def min_avg_window(times: list[int], k:int) -> float:
    if k <= 0 or k > len(times):
        raise ValueError("invalid k")
    
    window_sum = sum(times[:k]) #times[:2] is 0,1 (first 2 elements)
    best = window_sum
    
    for i in range(k, len(times)):  #range(2,5)->2,3,4 right open‚ö†Ô∏è
        print(f'now i is {i}')
        window_sum += times[i] - times[i - k]
        if window_sum < best:
            best = window_sum
        
        
    return window_sum

print(min_avg_window([30,40,25,35,20],2))
```

# boundry/index discipline
medium

```python
# Problem: Given a list of booleans representing whether a request failed, find the longest consecutive failure streak.

def longest_true_streak(flags: list[bool]) -> int:
    best = 0
    cur = 0
    
    for x in flags:
        if x:   #if it's True
            cur += 1
            if cur > best:
                best = cur
        else:
            cur = 0
    return best 


print(longest_true_streak([False, True, True, False, True, True, True]))  # 3
```

## deque
- we use deque() when we ‚ö†Ô∏è DON'T know the NUMBER of elements in the window
1. popleft(): removes and returns the leftmost (front) item from a deque

```python
from collections import deque

q = deque([1,2,3])
print(q.popleft())
print(q)
```

```python
# Problem: given request timestampds(seconds) and a limit L per 60-second window, return which requests are allowed.

from collections import deque

def allow_requests(times: list[str], limit: int) -> list[bool]:
    q = deque()
    out = []
    
    for t in times:
        # ‚ö†Ô∏è always compare the FIRST in the deque
        while q and q[0] <= t - 60:     #while q: while the deque is not empty
            q.popleft()
        if len(q) < limit:
            q.append(t)
            out.append(True)
        else:
            out.append(False)
    return out

print(allow_requests([0, 10, 20, 30, 40, 70, 71],3))
# eg, for t=70, the 60second window is [t-60,t] which is [10,70], which means any timestamp smaller than 10seconds should be removed because they are too old and expired
```

# **args & **kwargs
- 1. **args: Collect all extra positional arguments into a tuple named arg(xxx), not kv pairs
- 2. **kwargs: any keyword (kv) arguments

- this avoids: 1)New param is added;  2)Callers pass fewer params

```python
# **args and **kwargs example together

def log_event(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

log_event(
    "user_login",   #into args
    12345,  #into args  
    success=True,   #kwargs
    latency_ms=120  #kwargs
)
log_event
```

```python
# **args example

def aggregate_scores(*scores):
    print("Raw scores:", scores)

    if not scores:
        return None

    avg_score = sum(scores) / len(scores)
    return round(avg_score, 4)


result = aggregate_scores(0.82, 0.91, 0.88)
result
```

```python
# **kwargs example

def generate_answer(prompt, **gen_params):
    print("Prompt:", prompt)
    print("Generation params:", gen_params)

    temperature = gen_params.get("temperature", 0.7)
    max_tokens = gen_params.get("max_tokens", 256)

    return {
        "text": f"Generated answer for: {prompt}",
        "temperature": temperature,
        "max_tokens": max_tokens
    }

result = generate_answer(
    "Explain what RAG is",
    temperature=0.9,
    top_p=0.95,
    max_tokens=512
)
```

```python
# another example of LLM
def generate_answer(prompt, **gen_params):
    allowed = {"temperature", "top_p", "max_tokens"}

    for k in gen_params:
        if k not in allowed:
            raise ValueError(f"Unsupported param: {k}")

generate_answer("hello", temperature=0.8, batch=5)
```

# ‚≠êÔ∏è sliding window
- steps:
1) maintain a window with two pointers: "left" and "right"
2) expand "right" to include new characters
3) when the window satisfies the condition, try shrinking from "left" to make it minimal ‚≠êÔ∏è
4) keep track of the best window seen

- initial vars:
    - "left": we ALWAYS need "left"
    - "best": not mandatory. ONLY when we need to OPTIMIZE over all windows
    - ...

- commonly used for:
    - minimum window substring
    - longest substring without repeats
    - sum constraints
    - ...

- remember:
    - current window length: input[left, right+1] / best(left, right+1) / right - left + 1

- when to use "while":
    - Use while when the window may remain valid after shrinking once. ‚≠êÔ∏è
    - Use if (or single move) when shrinking once always breaks validity.

- difficulty:
    - easy: fixed-size windows(eg: max average subarray, simple counts)
    - medium: variable-size with constraints (eg: minimum window substring)
    - hard: multiple constraints, tricky invariants, or combining windows with other structures

## 1Ôºâlongest substring without repeating characters

```python
# self 20260204 ‚ùå

# edge case: 1)input is ""
def longest_substring(input: str)->int:
    # edge case handling
    if input == "":
        return -1
    
    left = input[0]
    store = input[0]
    longest = ""
    
    for index, i in enumerate(input[1:]):
        if i != left:
            store += i 
        else:
            longest = max(store, longest)
            store = i
    
    return longest

input = "abcabcbb"
longest_substring(input)
```

### solution

```python
"""  
EXPLANATION
1.  last[ch] >= left
        for example, "abba".
        when we process the second b, we already tossed the first a (index=0), and made left=last[ch]+1=1+1=2
        which means, when we process the second a(index=2), the old a(index=0) doesn't matter any more because the start of the window, which is the "left", is already 2!
"""

def longest_unique_substring_len(s: str) -> int:
    """ Keep a window that ALWAYS contains UNIQUE characters, expand it to the right, 
    and shrink it from the left when a duplicate appears. """
    last = {}   # a dict to keep: character ‚Üí last index where it appeared
    left = 0    # The start index of the current valid window / the earliest position of the valid substring
    best = 0    # stores the max length we've found so far
    for right, ch in enumerate(s):
        if ch in last and last[ch] >= left:     #MOST important line! If the previous ch is outside the window, it does NOT cause a conflict
            left = last[ch] + 1     # use 'abc a' as an example, here >= left means: The earliest valid substring must start after the old 'a'
        last[ch] = right    # update where we last saw this character
        best = max(best, right - left + 1)  # right-left+1: current window length
    return best

s = "abcabcbb"
print(longest_unique_substring_len(s))
```

## 2Ôºâminimum window containing all chars of t
(leetcode 76,medium)
- input: ("ADOBECODEBANC", "ABC")
- output: "BANC"

- use "while"

```python
# recommended

""" 
need tracks how far the window is from satisfying t,
missing == 0 means the window is valid,
shrinking continues until validity breaks.

eg:
s = "AABC"
t = "ABC"
step1: need[A]->1->0
step2: need[A]->0->-1   (we need "need[ch]" to remember the repeated chars so that later we can check whether we can shrink or not)
"""

from collections import Counter

def min_window(s: str, t: str) -> str:
    need = Counter(t)   # tracks how many of each character we still need
        # {"A":1, "B":1, "C":1}
        # ‚ö†Ô∏è Counter() sets missing keys default to 0 üòØ
    missing = len(t)    # total number of characters (including duplicates) still missing from the window
    left = 0    # define current window: s[left:right+1]
    best = (0, float("inf"))    #stores the best (smallest) valid window found so far
        # float("inf"): stands for positive infinite
        # later will be updated to :[left, window_length] ‚ö†Ô∏ènot right boundary
    
    # 1)expand the window
    for right, ch in enumerate(s):
        if need[ch] > 0:    # if i need this character
            missing -= 1
        need[ch] -= 1   # ‚ö†Ô∏è ‚ÄúOne ch has entered the window.‚Äù it's a FACT, regardless whether we need it or not.

        # 2) shrink left while still valid to make it minimal (because problem allows redunduncy, so we should do minimize ‚ö†Ô∏è)
        while missing == 0:
            """ 
            if we come here, it means:
            1. we have found all the chars we need, in the "t" (pattern)
            2. these chars are in the window -> [left, right+1]
            
            why while?
            for example, if the window is now : "AABBABC", we actually need to loop multiple times, to remove the first "A" and the first "B"
            """
            if right - left + 1 < best[1]:
                best = (left, right - left + 1)
                
            # try to shrink the window from left
            c = s[left]
            need[c] += 1    # stores the FACT
            if need[c] > 0:
                missing += 1
            left += 1   # update left boundary here

    start, length = best
    return "" if length == float("inf") else s[start:start+length]
```

```python
# self 20260205 reinforce the procedure (literally recited it..)‚úÖ
from collections import Counter
def find_min_window(input: str, pattern: str)->str:
    left = 0
    best = (0, float("inf"))
    need = Counter(pattern)
    missing = len(pattern)
    
    for right, ch in enumerate(input):
        # expand window
        if need[ch] > 0:
            missing -= 1
        need[ch] -= 1
        
        while missing == 0:
            # update window
            if right - left + 1 < best[1]:
                best = (left, right-left+1)
        
            # shrink from left
            c = input[left]
            need[c] += 1
            if need[c] > 0:
                missing += 1
            left += 1
    start, length = best
    return "" if input == "" else input[start:start+length] #not right-left+1 here


input = "ADOBECODEBANC"
pattern = "ABC"
find_min_window(input, pattern)
```

# longest consecutive sequence length
- input: [100,4,200,1,3,2]
- output: 4 (1,2,3,4)

```python
# recommended
def longest_consecutive(nums: list[int]) -> int:
    s = set(nums)   # remove duplicates because we don't need them; and also, this can do O(1) membership checks‚ùì
    best = 0
    for x in s:
        if x - 1 not in s:  # only start counting from numbers that are the BEGINNING of a sequence. if x-1 exsits, then x is in the middle of a sequence, so skip it
            y = x
            while y in s:
                y += 1
            best = max(best, y - x)
    return best
```

# prefix-sum + hashmap
- prefix-sum
    - it answers: ‚ÄúWhat is the sum from index 0 to index j?‚Äù (j=current index)

- hashmap
    - value ‚Üí "how many times this value has appeared so far"

## prefix-sum
standard format

```python
def generate_prefix_sum(nums):
    """ calculate the "SUM" for numbers before and including current index """
    prefix_sums = []
    current_sum = 0
    for num in nums:
        current_sum += num
        prefix_sums.append(current_sum)
    return prefix_sums

# Example usage:
nums = [3, 4, 1, 7, 9, 1]
p_sums = generate_prefix_sum(nums)
p_sums
# p_sums is [3, 7, 8, 15, 24, 25]
```

```python
# recommended

""" 
j: current index
i: some earlier index

three sums here:
    1) k: the sum from index i to index j
    2) s: prefix-sum from index 0 to index j
    3) x: the sum we need for substraction
    ------- s --------
       ------ k ------
    -x-
    0__i_____________j
    
    (-> x=s-k)

cnt[s - k]: ‚ÄúHow many earlier indices had prefix sum = s - k?‚Äù -> each corresponds to one VALID subarray

"""

from collections import defaultdict

def cnt_subarray(nums: list[int], k: int) -> int:
    cnt = defaultdict(int)  # records how many times each prefix sum has appeared so far.
    cnt[0] = 1  # 
    s = 0       # running prefix sum till current index
    ans = 0     # total count of subarrays summing to k
    for x in nums:
        s += x
        ans += cnt[s - k]   # if s-k never appears, it will be 0 because we've used defaultdict(int)
        cnt[s] += 1
    return ans

input = [3, 4, 7, -2, 2, 1, 4, 2]
k = 7
cnt_subarray(input, k)
```

# ‚≠êÔ∏è sliding window sum of size k
Return sums of each length-k window.
- input: [1,2,3,4],2
- output: [3,5,7]

solution A: seed-first window slide (classic)
- it starts with an initial value, intead of 0/empty
- more common

```python
# recommended ‚úÖ
def window_sums(nums: list[int], k: int) -> list[int]:
    # 0) guard clause
    if k <= 0 or k > len(nums):
        return []
    
    # 1) set seed state
    s = sum(nums[:k])
    # 2) emit first answer
    out = [s]
    for i in range(k, len(nums)):
        # 3) expand right, shrink left
        s += nums[i] - nums[i-k]
        # 4) emit result
        out.append(s)
    return out
```

solution B: build-up (with 0 and empty) then slide

```python
def window_sums(nums: list[int], k: int) -> list[int]:
    if k <= 0 or k > len(nums):
        return []
    out = []
    s = 0
    for i, x in enumerate(nums):
        # 1) expand window to the right
        s += x
        # 2) shrink window from the left
        if i >= k:
            s -= nums[i - k]          # nums[i - k] is the element that must be removed to keep the window size exactly k.
        # 3) record result if window is full
        if i >= k - 1:
            out.append(s)             # window is now full
    return out

# Example
print(window_sums([1, 2, 3, 4], 2))   # [3, 5, 7]
```

```python
# recommended
def merge_intervals(intervals: list[list[int]]) -> list[list[int]]:
    if not intervals:
        return []
    
    intervals.sort()
    out = [intervals[0]]
    for s, e in intervals[1:]:
        if s <= out[-1][1]:
            out[-1][1] = max(out[-1][1], e)
        else:
            out.append([s, e])
    return out
```

```python
# self 20260204

def merge_intervals(input: List[List[int]])->List[List[int]] | None:
    # suppose each List element has larger right than left
    
    """ 
    0) skip when x_right < x_left
    1) x_left <= curr_right, x_right >= curr_left -> overlap
    """
    curr = input[0]
    out = []
    
    for x in input[1:]:
        if x[1] < x[0]:
            return None
        if x[0] > curr[1] or x[1] < curr[0]:  # no overlap
            out.append(x)
        else:
            out.append([curr[0],x[1]])
        curr = x
        
    return out


input = [[1,3],[2,6],[8,10],[15,18]] 
merge_intervals(input)  #[[1,6],[8,10],[15,18]]
```

```python
# recommended

def rle_encode(s: str) -> str:
    if not s:
        return ""
    out = []     # store result
    cur = s[0]   # store current ‚≠êÔ∏è
    cnt = 1      # store count
    for ch in s[1:]:
        if ch == cur:
            cnt += 1
        else:
            out.append(f"{cur}{cnt}")
            cur, cnt = ch, 1
    out.append(f"{cur}{cnt}")
    return "".join(out)

input = "aaabbc"
rle_encode(input)
```

‚≠êÔ∏è set a "current pointer" is commonly used for traversing nested structures (dicts, trees, linked lists) to avoid repeatedly indexing from the root
- O(k) per use

```python
def get_nested(d: dict, keys: list, default=None):
    cur = d     # this sets a "current pointer" üòØ to the dict you're walking through. as the loop goes through keys, cur will be updated.
    for k in keys:
        if not isinstance(cur, dict) or k not in cur:
            return default
        cur = cur[k]
    return cur

a = {"a":{"b":2}}
keys = ["a","b"]
get_nested(a, keys)
```

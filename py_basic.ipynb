## basic unpacking
Unpack a fixed-length iterable into variables.
- ‚ö†Ô∏è Length must match exactly.

```python
# 1 to 1 example

tuple_a = (10,20)
x, y = tuple_a
print(x,y)

# but if we use 3 variable to hold the elements, it will be ValueError
# a,b,c = tuple_a #ValueError

# unpack list
list_a = [10,20]
x, y = list_a
print(x, y)

# unpack strings
s = "OK"
a, b = s
print(a,b)
# another string example
word = "Python"
first, *rest, last = word
print(first, rest, last)  # Output: P ['y', 't', 'h', 'o'] n
```

## bit_count()

## setdefault(): grouping (bucket)
(strong interview signal)
- mental model: grouping = create BUCKETS, then fill them
- RL intuition

```python
logs = [
    {"user": "A", "reward": 1},
    {"user": "A", "reward": 0},
    {"user": "B", "reward": 1}
]

group = {}
for log in logs:
    u = log["user"]
    group.setdefault(u,[]).append(log["reward"])
    """ setdefault()
    If u is already a key, return group[u].
    If u is missing, create group[u] = [] and return that list.
    """
group
```

## ‚≠êÔ∏è @classmethod & @staticmethod
- @classmethod receives cls (the class), often used for alternative constructors.
- @staticmethod is a namespaced function. it doesn‚Äôt get self or cls, and it doesn‚Äôt need an instance to run.

```python
class EmbeddingConfig:
    def __init__(self, model, dim):
        self.model = model
        self.dim = dim

    @classmethod
    def from_dict(cls, d):
        return cls(model=d["model"], dim=d["dim"])

    @staticmethod
    def supported_dims():
        return [384, 768, 1536]

cfg = EmbeddingConfig.from_dict({"model": "text-embed", "dim": 768})
print(cfg.model, cfg.dim)
print(EmbeddingConfig.supported_dims())
```

```python
# another example

class Temperature:
    def __init__(self, celsius: float):
        self.celsius = celsius

    def to_fahrenheit(self):
        # instance method: uses self (instance data)
        return self.celsius * 9 / 5 + 32

    @classmethod
    def from_fahrenheit(cls, f: float):
        # class method: constructs an instance (uses cls)
        return cls((f - 32) * 5 / 9)

    @staticmethod
    def is_valid(celsius: float):
        # static method: utility, no self/cls needed
        return celsius >= -273.15

print(Temperature.is_valid(-300))          # False (no instance needed)
t = Temperature(25)
print(t.to_fahrenheit())                   # 77.0 (uses instance)
print(Temperature.from_fahrenheit(32))     # Temperature instance
```

## @property
make an attribute look like a field while keeping rules inside.

```python
class LLMParams:
    def __init__(self, temperature):
        self.temperature = temperature

    @property
    def temperature(self):
        return self._temperature

    @temperature.setter     #rules for field/attribute "temperature"
    def temperature(self, value):
        if not (0 <= value <= 2):
            raise ValueError("temperature must be in [0, 2]")
        self._temperature = float(value)

p = LLMParams(0.7)
print(p.temperature)
# p.temperature = 3  # would raise
```

## use classA in classB

```python
class Retriever:
    def retrieve(self, query):
        return [f"doc about {query} (score=0.82)"]

class Generator:
    def generate(self, query, docs):
        return f"Answer to '{query}' using {len(docs)} docs."

class RAGPipeline:
    def __init__(self, retriever, generator):
        # use other 2 classes in this class
        self.retriever = retriever
        self.generator = generator

    def run(self, query):
        docs = self.retriever.retrieve(query)
        return self.generator.generate(query, docs)

pipe = RAGPipeline(Retriever(), Generator())
print(pipe.run("refund policy"))
```

# sliding window
easy to medium

```python
# Problem: Given daily delivery times, find the minimum average time over any consecutive window of size k.

def min_avg_window(times: list[int], k:int) -> float:
    if k <= 0 or k > len(times):
        raise ValueError("invalid k")
    
    window_sum = sum(times[:k]) #times[:2] is 0,1 (first 2 elements)
    best = window_sum
    
    for i in range(k, len(times)):  #range(2,5)->2,3,4 right open‚ö†Ô∏è
        print(f'now i is {i}')
        window_sum += times[i] - times[i - k]
        if window_sum < best:
            best = window_sum
        
        
    return window_sum

print(min_avg_window([30,40,25,35,20],2))
```

```python
# Problem: Given a list of booleans representing whether a request failed, find the longest consecutive failure streak.

def longest_true_streak(flags: list[bool]) -> int:
    best = 0
    cur = 0
    
    for x in flags:
        if x:   #if it's True
            cur += 1
            if cur > best:
                best = cur
        else:
            cur = 0
    return best 


print(longest_true_streak([False, True, True, False, True, True, True]))  # 3
```

```python
from collections import deque

q = deque([1,2,3])
print(q.popleft())
print(q)
```

```python
# Problem: given request timestampds(seconds) and a limit L per 60-second window, return which requests are allowed.

from collections import deque

def allow_requests(times: list[str], limit: int) -> list[bool]:
    q = deque()
    out = []
    
    for t in times:
        # ‚ö†Ô∏è always compare the FIRST in the deque
        while q and q[0] <= t - 60:     #while q: while the deque is not empty
            q.popleft()
        if len(q) < limit:
            q.append(t)
            out.append(True)
        else:
            out.append(False)
    return out

print(allow_requests([0, 10, 20, 30, 40, 70, 71],3))
# eg, for t=70, the 60second window is [t-60,t] which is [10,70], which means any timestamp smaller than 10seconds should be removed because they are too old and expired
```

```python
# **args example

def aggregate_scores(*scores):
    print("Raw scores:", scores)

    if not scores:
        return None

    avg_score = sum(scores) / len(scores)
    return round(avg_score, 4)


result = aggregate_scores(0.82, 0.91, 0.88)
result
```

```python
# **kwargs example

def generate_answer(prompt, **gen_params):
    print("Prompt:", prompt)
    print("Generation params:", gen_params)

    temperature = gen_params.get("temperature", 0.7)
    max_tokens = gen_params.get("max_tokens", 256)

    return {
        "text": f"Generated answer for: {prompt}",
        "temperature": temperature,
        "max_tokens": max_tokens
    }

result = generate_answer(
    "Explain what RAG is",
    temperature=0.9,
    top_p=0.95,
    max_tokens=512
)
```

# assert
- If the failure means ‚Äúthe programmer messed up‚Äù, use assert. (so usually xx_code == xxx we use assert because xx_code is defined by programmers/human. if the code is not correct, then the program will be broken)
- If the failure means ‚Äúthe user or environment messed up‚Äù, use raise or try...except.

- or easier: "should this never happen?" if yes, then assert ‚≠êÔ∏è

```python
# mix assert with others

def handle_query(query, index):
    assert isinstance(index, dict)   # developer guarantee(‚ÄúIf this is false, the program is already broken.‚Äù)

    if not query:
        raise ValueError("query is empty")   # user error

    try:
        return index[query]
    except KeyError:
        return []   # normal fallback
```

- invariant: An invariant is a condition that must always hold at a specific point in a program, and if it fails, it indicates a bug or violated system contract rather than a recoverable runtime error.

```python
# common asserts in AI/ML pipelines

def rerank(candidates, embeddings):
    assert len(candidates) > 0
    assert embeddings.shape[0] == len(candidates)
```
